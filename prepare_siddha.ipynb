```python
import pandas as pd
import glob
import os

# Пути
raw_dir = 'data/raw/siddha_watch/'
processed_dir = 'data/processed/'
os.makedirs(processed_dir, exist_ok=True)

# Соответствие активности -> число
activity_map = {'A': 0, 'B': 1, 'C': 2, 'D': 3, 'E': 4}

all_data = []

# Обработка каждого файла
for file_path in glob.glob(os.path.join(raw_dir, 'data_watch_*.txt')):
    # Извлекаем номер пользователя из имени файла
    user_id = int(file_path.split('_')[-1].split('.')[0])

    # Читаем CSV без заголовков
    df = pd.read_csv(
        file_path,
        sep=';',
        header=None,
        names=['activity', 'timestamp', 'gyro_x', 'gyro_y', 'gyro_z',
               'acc_x', 'acc_y', 'acc_z']
    )

    # Оставляем только нужные активности
    df = df[df['activity'].isin(activity_map.keys())].copy()

    # Заменяем буквы на числа
    df['activity'] = df['activity'].map(activity_map)

    # Добавляем колонку с ID пользователя
    df['user_id'] = user_id

    all_data.append(df)

# Объединяем все данные
final_df = pd.concat(all_data, ignore_index=True)

# Сохраняем результат
output_path = os.path.join(processed_dir, 'siddha_processed.csv')
final_df.to_csv(output_path, index=False)

print(f"Датасет сохранён в {output_path}, строк: {len(final_df)}")
```
